## Handling Outliers

```python
dfc_order_payment.info()
```

find outliers from payment_value

```python
q1 = dfc_order_payment['payment_value'].quantile(.25)
q3 = dfc_order_payment['payment_value'].quantile(.75)
iqr = q3-q1

UQ = q3+1.5*iqr
LQ = q1-1.5*iqr

print('Interval Quartile',iqr)
print('Upper Inteval',UQ)
print('Lower Interva',LQ)
```

Deciding to not drop the outliers data considering payment_value may have values more than **Upper Interval** values.

```python
dfc_order_payment = dfc_order_payment[(dfc_order_payment['payment_value']>=LQ)]
dfc_order_payment
```

```python
df_final_merged['purchase_year_month'] = df_final_merged['order_purchase_timestamp'].dt.strftime('%Y%m')
df_final_merged['purchase_year'] = df_final_merged['order_purchase_timestamp'].dt.year
df_final_merged['purchase_month'] = df_final_merged['order_purchase_timestamp'].dt.month
df_final_merged['purchase_month_name'] = df_final_merged['order_purchase_timestamp'].dt.month_name()
df_final_merged['purchase_day'] = df_final_merged['order_purchase_timestamp'].dt.day
df_final_merged['purchase_dayofweek'] = df_final_merged['order_purchase_timestamp'].dt.dayofweek
df_final_merged['purchase_day_name'] = df_final_merged['order_purchase_timestamp'].dt.day_name()
df_final_merged['purchase_hour'] = df_final_merged['order_purchase_timestamp'].dt.hour
```

## Numbers of Order per month

```python
#Calculate total order per month
df_order_permonth=df_final_merged.groupby('purchase_year_month')['order_id'].agg('count').reset_index()
df_order_permonth
```

```python
#Rename Collumn
df_order_permonth.columns = ['purchase_year_month','numbers_of_order']
df_order_permonth
```

```python
#Visualization
plt.figure(figsize=(20,10))
plt.title('Numbers of Order per month',fontsize=15)
sns.lineplot(x='purchase_year_month', y='numbers_of_order', data=df_order_permonth, color="coral", linewidth=3,
            markers=True)
```

*   The numbers of orders has **significantly increase** from September 2016 until August 2018.
*   There's a major **decrease of numbers of order** between **August 2018** and **September 2018**.
*   The Numbers of orders **reached its highest peak** on **November 2017**.



## Daily Order Trends (which day of week is the busiest)

```python
#daily order trends
df_order_perday=df_final_merged.groupby('purchase_day_name')['order_id'].agg('count').reset_index()
df_order_perday
```

```python
#Rename Collumn
df_order_perday.columns = ['purchase_day_name','numbers_of_order']
df_order_perday
```

```python
#Visualization
plt.figure(figsize=(15,10))
plt.title('Daily Order Trends by Day of',fontsize=15)
sns.barplot(x='purchase_day_name', y='numbers_of_order', data=df_order_perday, palette='gist_earth',
order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
```

**Monday** is the busiest day of week. Most of customer tend to do online shopping in **monday**, following by **tuesday**.

## Percentage of Orders per Payment Type

```python
df_order_payment=df_final_merged.groupby('payment_type')['order_id'].agg('count').reset_index()
df_order_payment
```

```python
#Rename Collumn
df_order_payment.columns = ['payment_type','numbers_of_order']
df_order_payment
```

```python
#Visualization
plt.figure(figsize=(15,10))
plt.title('% Orders per Payment Type',fontsize=15)
plt.pie(df_order_payment['numbers_of_order'],labels=df_order_payment['payment_type'],autopct='%1.1f%%')

plt.show()
```

Most of customers preferred to pay with **credit_card**, followed by **boleto**.

## Total Customer per State

```python
df_customer_perstate=df_final_merged.groupby('customer_state')['customer_id'].agg('count').reset_index()
df_customer_perstate
```

```python
#Rename Collumn
df_customer_perstate.columns = ['customer_state','total_customer']
df_customer_perstate
```

```python
#Sort total customer by descending state
df_customer_perstate.copy()
df_customer_perstate.sort_values(by='total_customer',ascending=False,inplace=True)
df_customer_perstate
```

```python
#Visualization
plt.figure(figsize=(15,10))
plt.title('Total Customer per State',fontsize=15)
sns.barplot(x='total_customer', y='customer_state', data=df_customer_perstate, palette='gist_earth', orient='h', ci=None)
```

Most of customers come from **Sau Paulo** state, while
**Roraima** has the least number of customer.
